```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("first_dose_data_transpose.csv")
```

```python
df.iloc[[1]] 
```

```python
new = df.iloc[1:-1].copy()
new
```

```python
old = df.iloc[2:].copy()
old
```

```python
for i in range(1,181):
  df.iloc[i+1,1:] = old.iloc[i-1,1:] - new.iloc[i-1,1:]
```

```python
df.to_csv('first_dose_daily_increase.csv')
```

```python
df = pd.read_csv("second_dose_data_transpose.csv")
new = df.iloc[1:-1].copy()
old = df.iloc[2:].copy()
for i in range(1,181):
  df.iloc[i+1,1:] = old.iloc[i-1,1:] - new.iloc[i-1,1:]
df.to_csv('second_dose_daily_increase.csv')
```

```python
df = pd.read_csv("booster_dose_data_transpose.csv")
new = df.iloc[1:-1].copy()
old = df.iloc[2:].copy()
for i in range(1,127):
  df.iloc[i+1,1:] = old.iloc[i-1,1:] - new.iloc[i-1,1:]
df.to_csv('booster_dose_daily_increase.csv')
```
